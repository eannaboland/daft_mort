```python
from statsmodels.formula.api import ols
from daftlistings import Daft, SaleType, Request
import matplotlib.pyplot as plt
import time
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
offset = 0
df = pd.DataFrame()
```

```python

while True:
    daft = Daft()
    daft.set_county("Dublin City")
    daft.set_offset(offset)
    daft.set_listing_type(SaleType.PROPERTIES)

    listings = daft.search()

    if not listings:
        break

    for listing in listings:
        time.sleep(np.round((np.random.rand(1)/5)[0],3))
        d = Request().get(listing.daft_link)
        try:
            area = d.find('div', {'class':"PropertyOverview__propertyOverviewDetails"}).contents[4].replace(' ', '').replace('\n','').replace('m', '')
            print(area)
            
        except:
            area = ''
        data = pd.DataFrame([listing.price,  listing.bedrooms,  listing.bathrooms, listing.dwelling_type, listing.town, listing.county ,listing.daft_link, area]).T
        df = df.append(data)
    print(offset)
    print(df.shape)
    offset += 20
    time.sleep(np.round((np.random.rand(1)/5)[0],3))
```

```python
df.columns = ["Price", 'Beds', 'Baths', 'Type', 'jlcity', 'jlregion1', 'URL', 'Area']
```

```python
df = df[~df.Price.str.contains("Price")].copy()
```

```python
df.Price = df.Price.str.replace('', '').str.replace(',','').str.extract('(\d+)').astype(float)
```

```python
df.Beds = df.Beds.str.replace('N/A', '99').str.replace('+', '').astype(int)
df = df[df.Beds!=99]
```

```python
df.Baths = df.Baths.str.replace('N/A', '99').str.replace('+', '').astype(int)
df = df[df.Baths!=99]
```

```python
df['Rooms'] = df.Beds + df.Baths
```

```python
#df['Price Per Bedroom'] = df.Price/df.Beds
#df['Price Per Bathroom'] = df.Price/df.Baths
#df['Price Per Bathroom and Bedrooms'] = df.Price/(df.Baths + df.Beds)
```

```python
df['beds_baths'] = df.Beds*df.Baths
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('data_' + str(pd.to_datetime('today').date()) + '.csv')
```

```python
df.loc[df.Area =='', 'Area'] = 0 #.astype(float)
```

```python
df.Area = df.Area.astype(float)
```

```python
df = df[df.Area!=0].copy()
```

```python
df = df[df.jlcity.isin(df.jlcity.value_counts().loc[df.jlcity.value_counts()>1].index)].copy()
df = df[~df.isna().any(axis=1)].copy()
```

```python
fit = ols('Price ~ C(Type) + C(jlregion1) +C(jlcity) + Beds + Baths + Area', data=df).fit() 
```

```python
df['model'] = fit.fittedvalues
```

```python
df['residual'] = df.Price - df.model
df['pdif'] = (df.Price - df.model)/df.Price
```

```python
import requests
from bs4 import BeautifulSoup

class Request:
    def __init__(self, verbose=False):
        self._headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0'}
        self._verbose = verbose

    def get(self, url, params=None):
        req = requests.get(url, headers=self._headers, params=params)
        if self._verbose:
            logger.info("URL: " + req.url)
            logger.info("Status code: " + str(req.status_code))

        if req.status_code != 200:
            raise DaftException(reason=req.reason)

        soup = BeautifulSoup(req.content, 'html.parser')
        return soup

    def post(self, url, params=None):
        req = requests.post(url, params=params, headers=self._headers)

        if self._verbose:
            logger.info("URL: " + req.url)
            logger.info("Status code: " + str(req.status_code))

        if req.status_code != 200:
            raise DaftException(reason=req.reason)

        return req
```
