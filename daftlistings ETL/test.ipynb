```python
df.columns = ["Price", 'Beds', 'Baths', 'Type', 'jlcity', 'jlregion1']
```

```python
df = df[~df.Price.str.contains("Price")]
```

```python
df['AMV_flag'] = df.Price.str.contains('amv')
```

```python
df.Price = df.Price.str.replace('€', '').str.replace(',','').str.extract('(\d+)').astype(float)
```

```python
df.Beds = df.Beds.str.replace('N/A', '99').str.replace('+', '').astype(int)
df = df[df.Beds!=99]
```

```python
df.Baths = df.Baths.str.replace('N/A', '99').str.replace('+', '').astype(int)
df = df[df.Baths!=99]
```

```python
df['Rooms'] = df.Beds + df.Baths
```

```python
df['Price Per Bedroom'] = df.Price/df.Beds
df['Price Per Bathroom'] = df.Price/df.Baths
df['Price Per Bathroom and Bedrooms'] = df.Price/(df.Baths + df.Beds)
```

```python
plt.figure(figsize=(20,10))
#plt.xscale('log')
for typ in df.Type.value_counts()[0:2].index:#, 'Apartment for sale']:
    plt.hist(df[df["Type"] == typ]["Price"].reset_index(drop=True), alpha=0.6, label=typ)

plt.suptitle('Dublin 4 Housing Data - Price Per (Bed + Bath) In Property', fontsize=20)
plt.xlabel('Price', fontsize=18)
plt.ylabel('Quantity', fontsize=16)
plt.legend()
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

class Request:
    def __init__(self, verbose=False):
        self._headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0'}
        self._verbose = verbose

    def get(self, url, params=None):
        req = requests.get(url, headers=self._headers, params=params)
        if self._verbose:
            logger.info("URL: " + req.url)
            logger.info("Status code: " + str(req.status_code))

        if req.status_code != 200:
            raise DaftException(reason=req.reason)

        soup = BeautifulSoup(req.content, 'html.parser')
        return soup

    def post(self, url, params=None):
        req = requests.post(url, params=params, headers=self._headers)

        if self._verbose:
            logger.info("URL: " + req.url)
            logger.info("Status code: " + str(req.status_code))

        if req.status_code != 200:
            raise DaftException(reason=req.reason)

        return req
```
